# 빵셀러 데이터 정제·집계 보고서

## 1. 데이터 개요
- 출처: (가상) 빵셀러 베이커리 체인 운영 로그
- 기간: 2025-04-01 ~ 2025-06-30 (3개월)
- 원본 행 수 / 정제 후 행 수: 2164행 → 1784행

## 2. 발견한 품질 문제
- 결측: revenue에 60개 결측값이 있다. (2.77%)
- 중복: 321건
- 이상치: quantity에 14.56%(원인: 50이라는 너무 큰 숫자가 있기 때문), revenue에 17.30%(음수값이 하나 존재한다.) 있의 이상치가 존재한다. 
- 표기 혼재: 케이크/케익 에 표기혼재가 있다. 

## 3. 처리 결정과 근거 (5줄로)
1. 중복 행 321건 → df.drop_duplicates()를 사용해서 중복 행들을 모두 제거함
2. store_id 공백 → stripe를 사용해서 공백을 모두 제거함
3. item '케익' → '케이크' → replace를 적용하여 두 값을 모두 케이크로 통일시켰다.
4. revenue 음수 → 별도 보관 1건 (환불과 관련된 수치일 수도 있으나, 지금 여기에서는 환불이 중점은 아니기 때문에 일단 별도로 보관한다.)
5. revenue 결측 60건 → 제거 (이유: 결측치의 비율이 크지 않고, 날짜형을 적용한 date_str로 정렬하고나서 msno를 사용해서 그래프를 확인해보니 시간과의 관계도 따로 보이지 않기 때문에 삭제하는 것으로 결정하였다. 

## 4. 주요 KPI 결과 (2줄)
- (월·매장 관점) 가장 매출이 큰 매장과 그 월: S06, 2025-04
- (상품 관점) 매출 1위 상품: 쿠키

2. 매장/상품 단위 KPI만 봐서, "왜" S06이 1위인지(입지, 유동인구, 프로모션 등) 원인 분석은 안 될 것 같다.

2. revenue 음수 1건 → 환불 여부 실제 확인
